```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,columns=['國文','英文','數學','地理','歷史'],
             index = range(1,51)
             )
scores_pd.index.name = "學號"
scores_pd.columns.name = "科目"
scores_pd
scores_pd['國文'] + scores_pd['英文'] 
+ scores_pd['地理'] + scores_pd['數學'] + scores_pd['歷史']
```

```python
scores_pd['總分'] = scores_pd[['國文','英文','數學','地理','歷史']].sum(axis=1)
scores_pd
scores_pd['平均'] = scores_pd[['國文','英文','數學','地理','歷史']].mean(axis=1)
scores_pd
```

```python
scores_pd['排名']= scores_pd['總分'].rank(method='min',ascending=False)
scores_pd.sort_values(by='排名')
```

```python
scores_pd.style\
.highlight_between(left=0,
                    right=59,
                    color='blue',
                    subset=['國文','英文','數學','地理','歷史'],
                    props='color:red:background-color:white')\
.format(precision=0,subset=['平均','排名'])
```
