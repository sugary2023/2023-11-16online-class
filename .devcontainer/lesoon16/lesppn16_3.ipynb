```python
import pandas as pd
import matplotlib.pyplot as plt
scores = [
    {'chinese':98,'english':93,'math':92,'history':95,'discover':100},
    {'chinese':93,'english':100,'math':95,'history':92,'discover':97}
]
dataFrame = pd.DataFrame(index=["Cody","Gary"],data=scores)
display(dataFrame.T)
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
dataFrame.T.plot(ax=axes,style=['--','-'],marker='o')
axes.set_title('王暄睿和蘇子滕的分數表',fontproperties=font)
axes.grid(visible=True,axis='y')
plt.show()
```
